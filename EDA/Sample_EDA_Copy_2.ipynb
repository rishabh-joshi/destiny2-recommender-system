We can get the fireteamIDs and characterIDs of every player in every match


```python
ids = dict()
fireteamids = dict()
i = 0
for mtch in match_data:
    ids[i] = []
    fireteamids[i] = []
    for player in mtch["Response"]["entries"]:
        fireteamids[i].append(player["values"]["fireteamId"]["basic"]["value"])
        ids[i].append(player["characterId"])
    i = i+1
print(ids)
print(fireteamids)
```

```python
full_list = [teamids for j in fireteamids for teamids in fireteamids[j]]
print(len(set(full_list))) #Total number of fireteamids across all matches - includes repeated 0s for no fireteam
```

```python
teamsets = list()
for j in fireteamids: #matches
    teamsets.append([])
    team1set = set()
    team2set = set()
    for k in range(4):
        team1set.add(ids[j][k])
    for k in range(4,8):
        team2set.add(ids[j][k])
    
    teamsets[j].append(team1set)
    teamsets[j].append(team2set)

print(teamsets)
```

```python
allsets = [frozenset(tmset) for mtchsets in teamsets for tmset in mtchsets]
print(allsets)
```

The above indicates that the fireteamID may not be directly representative of the team composition, since there are 29 different teams in terms of the characterIDs they are composed of, but 25 different fireteamIDs. However since we are only looking at full teams it would be better to check only among those.

```python
fullteam1matches = []
fullteam2matches = []
fullteam1fireteamid = []
fullteam2fireteamid = []
mtchteam1set = {}
mtchteam2set = {}
for mtch in fireteamids:
    team1set = set()
    team2set = set()
    chk = 0
    for k in range(4):
        team1set.add(fireteamids[mtch][k])
    for k in range(4,8):
        team2set.add(fireteamids[mtch][k])
    if len(team1set) ==1:
        fullteam1matches.append(mtch)
        fullteam1fireteamid.append(fireteamids[mtch][0])
        mtchteam1set[mtch] = team1set
        
        
    if len(team2set) ==1 :
        fullteam2matches.append(mtch)
        fullteam2fireteamid.append(fireteamids[mtch][5])
        mtchteam2set[mtch] = mtchteam2set
        
fullteam1matches = np.array(fullteam1matches)
fullteam2matches = np.array(fullteam2matches)
fullteam1fireteamid = np.array(fullteam1fireteamid)
fullteam2fireteamid = np.array(fullteam2fireteamid)
```

```python
print(np.array(fullteam1matches))
print(fullteam1matches)
print(set(fullteam1fireteamid))
```

```python
#allfullteammatches
unequalteam = {}
for fullteam1 in set(fullteam1fireteamid):
    i = 1
    unequalteam[fullteam1] = 0
    
    for mtch in fullteam1matches[fullteam1fireteamid == fullteam1]:
        
        if i == 1:
            chkset = mtchteam1set[mtch]
        if chkset != mtchteam1set[mtch]:
            unequalteam[fullteam1] = unequalteam[fullteam1] + 1
    
    for mtch in fullteam2matches[fullteam2fireteamid == fullteam1]:
        
        if i == 1:
            chkset = mtchteam2set[mtch]
        if chkset != mtchteam2set[mtch]:
            unequalteam[fullteam1] = unequalteam[fullteam1] + 1

print(unequalteam)
        
        
    
```

We can see that the composition of characterIDs remains the same in all matches where all 4 players in a team belong to one fireteamID, for the sample dataset.
Therefore if this extends to the universe of PvP matches, and if characterID uniquely represents a unique player's specific character, then the fireteamID can be used to track a team across all its matches.

```python
import json
import pandas as pd
import pprint
```

Loading sample data

```python
with open("PGCR.json") as json_file: #"Data/Destiny2_allPvP_matches/PGCR_graph_dataset0.json
    match_data = json.load(json_file)
    print(match_data)
```

```python
match_data[0].keys() #keys inside each match
```

We can get the fireteamIDs and characterIDs of every player in every match


```python
ids = dict()
fireteamids = dict()
i = 0



for mtch in match_data:
    ids[i] = []
    fireteamids[i] = []
    for player in mtch["Response"]["entries"]:
        fireteamids[i].append(player["values"]["fireteamId"]["basic"]["value"])
        ids[i].append(player["characterId"])
    i = i+1
print(ids)
print(fireteamids)
```

```python
full_list = [teamids for j in fireteamids for teamids in fireteamids[j]]
print(len(set(full_list))) #Total number of fireteamids across all matches - includes repeated 0s for no fireteam
```

```python
print([str(item) + ":" + str(full_list.count(item)) for item in set(full_list)]) 
#Appearances of each fireteamID (no. of players)
```

```python
print([str(len(set(fireteamids[k]))) for k in fireteamids])
#No. of fireteamIDs in each match
```

```python
teamsets = list()
for j in fireteamids: #matches
    teamsets.append([])
    team1set = set()
    team2set = set()
    for k in range(4):
        team1set.add(ids[j][k])
    for k in range(4,8):
        team2set.add(ids[j][k])
    
    teamsets[j].append(team1set)
    teamsets[j].append(team2set)

print(teamsets)
```

```python
allsets = [frozenset(tmset) for mtchsets in teamsets for tmset in mtchsets]
print(allsets)
```

```python
setofsets = set(allsets)
print(str(len(setofsets)))
```

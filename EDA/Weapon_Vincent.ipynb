##  itemType, itemTypeAndTierDisplayName, itemTypeDisplayName, loreHash

```python
'''
 'itemType': 3,
 'itemTypeAndTierDisplayName': 'Legendary Sniper Rifle',
 'itemTypeDisplayName': 'Sniper Rifle',
 'loreHash': 174804903,
'''
itemType_list=[]
itemTypeAndTierDisplayName_list=[]
itemTypeDisplayName_list=[]
loreHash_list=[]
for reference_id in weapon_data0:
    if weapon_data0[reference_id].get('itemType') ==None:
        itemType_list.append(None)
    else:
        itemType = weapon_data0[reference_id]['itemType']
        itemType_list.append(itemType)
        
    if weapon_data0[reference_id].get('itemTypeAndTierDisplayName') ==None:
        itemTypeAndTierDisplayName_list.append(None)
    else:
        itemTypeAndTierDisplayName = weapon_data0[reference_id]['itemTypeAndTierDisplayName']
        itemTypeAndTierDisplayName_list.append(itemTypeAndTierDisplayName)
        
    if weapon_data0[reference_id].get('itemTypeDisplayName') ==None:
        itemTypeDisplayName_list.append(None)
    else:
        itemTypeDisplayName = weapon_data0[reference_id]['itemTypeDisplayName']
        itemTypeDisplayName_list.append(itemTypeDisplayName)
        
    if weapon_data0[reference_id].get('loreHash') ==None:
        loreHash_list.append(None)
    else:
        loreHash = weapon_data0[reference_id]['loreHash']
        loreHash_list.append(loreHash)
    
    
```

```python
print(len(itemType_list))
print(len(itemTypeAndTierDisplayName_list))
print(len(itemTypeDisplayName_list))
print(len(loreHash_list))
```

## Levels 

```python

#each weapon has lots of "levels" from different comp stats
level_list=[]
for reference_id in weapon_data0:
    
    if weapon_data0[reference_id].get('sourceData') ==None:
        level_list.append(None)
    else:
        level_inner=[]
        for i in weapon_data0[reference_id]['sourceData']['sources']:
            level_inner.append(i['level'])  
        level_list.append(level_inner)
```

```python
len(level_list)
```

```python
level_list[0:5]
```

```python
weapon_table = pd.DataFrame(
    {'reference_id': reference_id_list,
     'itemTypeDisplayName':itemTypeDisplayName_list,
     'itemTypeAndTierDisplayName':itemTypeAndTierDisplayName_list,
     'weapon_name': weapon_name_list,
     'equipmentSlotTypeHash':equipmentSlotTypeHash_list,
     'tierTypeName':tierTypeName_list,
     'itemCategoryHashes':itemCategoryHashes_list,
     'itemSubType':itemSubType_list,
     'itemType':itemType_list, 
     'loreHash_list':loreHash_list,
     'levels':level_list
    })
```

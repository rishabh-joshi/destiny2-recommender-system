```python
import pandas as pd
import numpy as np
import matplotlib.pylab as plt
```

```python
t2 = pd.read_csv("../Data/Updated_Tables/table2_new_teamid.csv")
```

```python
t6 = pd.read_csv("../Data/Updated_Tables/table6_complete.csv")
```

```python
t2.head() #Primary Key is teamID,matchID,playersID
```

```python
t6.head() #Timestamp for every possible match
```

```python
merged = t2.merge(t6, left_on = "matchID", right_on = "MatchID") #Add timestamps to matches.
```

```python
merged.head()
```

```python
merged = merged.sort_values(by=["Date", "Time"], ascending=False) #Sorting by time
```

```python
merged.head(10)
```

```python
temp = merged
```

### Converting from Player-Match key to Team-Match by dropping duplicates

```python
temp_uniq = temp.drop_duplicates(subset = ["MatchID", "teamID"])
```

```python
temp_uniq = temp_uniq.sort_values(by = ["Date", "Time"]) #Sorting according to time again
```

```python
temp_uniq.head()
```

### Adding Match Number, sorted according to time

```python
temp_uniq["MatchNum"] = temp_uniq.groupby("teamID").cumcount()+1
```

```python
temp_uniq.head()
```

## SET NUMBER OF MATCHES USED TO CALCULATE WIN RATE BELOW

```python
MatchWindow = 15
```

### Filtering to only consider relevant matches

```python
temp_uniq_filtered = temp_uniq[temp_uniq.MatchNum < MatchWindow + 1]
```

### Calculating Win Ratio as mean of Wins column

```python
temp_uniq_filtered["Wins"] = 1 - temp_uniq_filtered["standing"]
```

```python
FinalDF = temp_uniq_filtered[['teamID', 'Wins']]
```

```python
FinalTable = FinalDF.groupby("teamID").mean().reset_index()
```

```python
FinalTable.describe()
```

```python
FinalTable.to_csv("../Data/Updated_Tables/WinRatio_upto15.csv")
```

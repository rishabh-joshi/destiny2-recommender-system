```python
# remove fireteamID = 0
print(merged.shape)
merged = merged[merged["fireteamID"] != 0]
print(merged.shape)
```

```python
# Add the skills to compute the learning rate for in this list
skills = ["Grenade", "Melee", "Super", "assists", "avg_score_per_life", "KDA", "efficiency"]
columns = ["playersID", "fireteamID", "MatchID", "Date", "Time"]
columns.extend(skills)
merged = merged[columns]
```

```python
merged = merged.sort_values(by=["Date", "Time"])
merged.head()
```

```python
from sklearn.linear_model import LinearRegression
def model(df):
    y = df[[skills[0]]].values
#     if y.shape[0] <= 1 : return [0,0]
    X = np.array(range(y.shape[0])).reshape(-1,1)
    return [LinearRegression().fit(X, y.reshape(-1,1)).coef_[0][0] for y in df[skills].T.values]
```

```python
learning_rates = merged.groupby(["playersID", "fireteamID"]).apply(model)
```

```python
learning_rates = learning_rates.reset_index()
```

```python
lr_subset = pd.DataFrame(learning_rates.iloc[:,2].values.tolist(), columns = ["lr_" + skill for skill in skills])
```

```python
id_info = learning_rates[["playersID", "fireteamID"]]
```

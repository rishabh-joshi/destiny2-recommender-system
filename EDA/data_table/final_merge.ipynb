## Reformat tables 

```python
#table1
table1=table1[["matchID","mode","standing","team_id","teamscore"]]
table1.head()
```

```python
#table2
table2=table2.drop('Unnamed: 0',axis=1)
```

```python
#table4
table4=table4[["reference_id","weapon_name",'itemTypeDisplayName',"itemTypeAndTierDisplayName","tierTypeName"]]
table4.head()
```

```python
table5=table5[['matchID','playersID', 'referenceID', 'uniqueWeaponKills',
       'uniqueWeaponKillsPrecisionKills', 'uniqueWeaponPrecisionKills']]
table5.head()
```

## Merge 

### First, merge table 5 and table 4 by reference id 

```python
table4_5=pd.merge(table5,table4,how="left",left_on="referenceID",right_on="reference_id")
table4_5.head(20)#primary key = playerid+matchid+referenceid
```

```python
type(table4_5["playersID"])
```

### Merge 4_5 with table 2 

```python
#key = match id +player id
table2_4_5=pd.merge(table4_5,table2, how="inner", left_on=["matchID","playersID"],right_on=["matchID","playersID"])
table2_4_5.head(20)
```

### merge 2_4_5 with table1 

```python
table1_2_4_5=pd.merge(table2_4_5,table1,how="inner",left_on="matchID",right_on="matchID")
table1_2_4_5.head(20)
table1_2_4_5.columns
```

### Remove Reduancy Attribute

```python
table1_2_4_5=table1_2_4_5.drop('reference_id',axis=1)
```

```python
table1_2_4_5.head()
```

### Merge 1_2_4_5 with table 3 

```python
table1_2_3_4_5=pd.merge(table1_2_4_5,table3,how="inner",left_on="playersID",right_on="playerID")
table1_2_3_4_5.head()
```

```python
table1_2_3_4_5=table1_2_3_4_5.drop("playerID",axis=1)
```

**Creating vectors**

```python
def teamprofile(df):
    df["values"] = 1
    table_per = df.pivot_table(index="teamID", columns="performance_cluster", values = "values",aggfunc=sum)
    table_weapon = df.pivot_table(index="teamID", columns="weapon_pro_cluster", values = "values",aggfunc=sum)
    table_per = table_per.fillna(0)
    table_weapon = table_weapon.fillna(0)
    finalresult = pd.concat([table_per, table_weapon], axis=1).reset_index()
    return finalresult
```

```python
finalresult = teamprofile(teamset)
```

**Calculate the distance matrix using city block**

```python
dist2 = pd.DataFrame(squareform(pdist(finalresult.iloc[:, 1:],metric = 'cityblock')), 
                     columns=finalresult.teamID.unique(), index=finalresult.teamID.unique()).reset_index()
```

**Find the closest teams**

```python
def find_min(dist):
    teamnames = []
    topteam = []
    for i in range(0,len(dist)):
        m = min(dist.iloc[i,1:])
        col_indices = np.where(dist.iloc[i,1:] == m)
        col_values = dist.columns[col_indices]
        teamnames.append(dist.iloc[i,0])
        topteam.append(col_values)
    finaltable = pd.DataFrame(       
    {'teamid': teamnames,
     'topteam': topteam,
    })
    return finaltable
        
```

```python
finaltable = find_min(dist2)
```
